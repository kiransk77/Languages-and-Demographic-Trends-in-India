```python
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
#select list of columns required in c8
c8="DDW-0000C-08.xlsx"
cols_c8=[1,3,4,5,10,11,16,17,19,20,22,23,25,26,28,29,31,32,34,35,37,38,40,41]
#load c8
df_c8=pd.read_excel(c8,usecols=cols_c8,skiprows=[0,2,3,4,5,6],index_col=False)
df_c8
```

```python
#rename the columns
df_c8.rename(columns={'State':'State Code',
                      'Area Name':'Name',
                      'Unnamed: 10':'Males_Illiterate',
                      'Unnamed: 11':'Females_Illiterate',
                      'Unnamed: 16':'Males_Literate without education level',
                      'Unnamed: 17':'Females_Literate without education level',
                      'Unnamed: 19':'Males_Below Primary',
                      'Unnamed: 20':'Females_Below Primary',
                      'Unnamed: 22':'Males_Primary',
                      'Unnamed: 23':'Females_Primary', 
                      'Unnamed: 25':'Males_Middle',
                      'Unnamed: 26':'Females_Middle',
                      'Unnamed: 28':'Males_Matric/Secondary',
                      'Unnamed: 29':'Females_Matric/Secondary',
                      'Unnamed: 31':'Males_Higher secondary/Intermediate/Pre-University/Senior secondary',
                      'Unnamed: 32':'Females_Higher secondary/Intermediate/Pre-University/Senior secondary',
                      'Unnamed: 34':'Males_Non-technical diploma or certificate not equal to degree',
                      'Unnamed: 35':'Females_Non-technical diploma or certificate not equal to degree',
                      'Unnamed: 37':'Males_Technical diploma or certificate not equal to degree',
                      'Unnamed: 38':'Females_Technical diploma or certificate not equal to degree',
                      'Unnamed: 40':'Males_Graduate and above',
                      'Unnamed: 41':'Females_Graduate and above',},inplace=True)

#df_c8.drop(['    Total','Illiterate','Literate','Unnamed: 42', 'Unnamed: 45'],axis=1,inplace=True)
```

```python
#preprocess c8 for req data
df_c8=df_c8[(df_c8['Total/']=='Total')&(df_c8['Age-group']=='All ages')]
df_c8.drop(['Total/','Age-group'],axis=1,inplace=True)
df_c8
```

```python
#adjusting the literacy columns as per those in c19
df_c8['Males_Literate but below primary']=df_c8['Males_Literate without education level']+df_c8['Males_Below Primary']
df_c8['Males_Matric/Secondary but below graduate']= df_c8['Males_Matric/Secondary']+df_c8['Males_Higher secondary/Intermediate/Pre-University/Senior secondary']+df_c8['Males_Non-technical diploma or certificate not equal to degree']+df_c8['Males_Technical diploma or certificate not equal to degree']

df_c8['Females_Literate but below primary']=df_c8['Females_Literate without education level']+df_c8['Females_Below Primary']
df_c8['Females_Matric/Secondary but below graduate']= df_c8['Females_Matric/Secondary']+df_c8['Females_Higher secondary/Intermediate/Pre-University/Senior secondary']+df_c8['Females_Non-technical diploma or certificate not equal to degree']+df_c8['Females_Technical diploma or certificate not equal to degree']

df_c8.rename(columns={'Males_Primary':'Males_Primary but below middle','Females_Primary':'Females_Primary but below middle','Males_Middle':'Males_Middle but below matric/secondary','Females_Middle':'Females_Middle but below matric/secondary'},inplace=True)
df_c8['State Code']=df_c8['State Code'].astype(int)
```

```python
#separating male and female data from c8
df_c8_m=df_c8[['State Code','Name','Males_Illiterate','Males_Literate but below primary','Males_Primary but below middle','Males_Middle but below matric/secondary','Males_Matric/Secondary but below graduate','Males_Graduate and above']]
df_c8_f=df_c8[['State Code','Name','Females_Illiterate','Females_Literate but below primary','Females_Primary but below middle','Females_Middle but below matric/secondary','Females_Matric/Secondary but below graduate','Females_Graduate and above']]

df_c8_m.columns=['State Code','Name','Illiterate','Literate but below primary','Primary but below middle','Middle but below matric/secondary','Matric/Secondary but below graduate','Graduate and above']
df_c8_f.columns=['State Code','Name','Illiterate','Literate but below primary','Primary but below middle','Middle but below matric/secondary','Matric/Secondary but below graduate','Graduate and above']

df_c8_f.reset_index(drop=True,inplace=True)
df_c8_m.reset_index(drop=True,inplace=True)
df_c8_m
```

```python
#converting rows into columns
df_c8_m = pd.melt(df_c8_m,id_vars=['State Code','Name'],
        var_name='Educational_level', value_name='TOT_M')
df_c8_m.sort_values(by="State Code",inplace=True,ignore_index=True)

df_c8_f = pd.melt(df_c8_f,id_vars=['State Code','Name'],
        var_name='Educational_level', value_name='TOT_F')
df_c8_f.sort_values(by="State Code",inplace=True,ignore_index=True)
```

```python
#merging the earlier separated male and female data into single data frame
cols_merge = ['State Code','Name','Educational_level']
df_c8 = pd.merge(df_c8_m,df_c8_f,how='inner',left_on=cols_merge,right_on = cols_merge)
df_c8.drop(['Name'],axis=1,inplace=True)
df_c8['State Code']=df_c8['State Code'].astype(int)
df_c8
```

```python
#load c19
c19="DDW-C19-0000.xlsx"
cols=[0,3,4,6,7,9,10]
col_names=["State Code","Total/","Educational_level","Males_2ndL","Females_2ndL","Males_3rdL","Females_3rdL"]
df_c19=pd.read_excel(c19,usecols=cols,names=col_names,skiprows=[0,2,3,4,5])
df_c19
```

```python
#selecting required literacy groups 
exl=['Total','Literate']
df_c19=df_c19[(df_c19['Total/']=='Total')&(~df_c19.Educational_level.isin(exl))]
df_c19.drop(['Total/'],axis=1,inplace=True)
df_c19['State Code']=df_c19['State Code'].astype(int)
df_c19.reset_index(drop=True,inplace=True)
df_c19
```

```python
#merge c8 and c19
df_merge=pd.merge(df_c8,df_c19,how='inner',left_on=['State Code','Educational_level'],right_on = ['State Code','Educational_level'])

#cal one lang speakers of males and females
df_merge["Males_1stL"]=df_merge["TOT_M"]-df_merge["Males_2ndL"]
df_merge["Females_1stL"]=df_merge["TOT_F"]-df_merge["Females_2ndL"]

#cal 2 lang speakers of males and females 
df_merge["Males_2ndL"]=df_merge["Males_2ndL"]-df_merge["Males_3rdL"]
df_merge["Females_2ndL"]=df_merge["Females_2ndL"]-df_merge["Females_3rdL"]
df_merge
```

```python
#cal ratios of one lang, two lang and three lang speakers of males
df_merge["ratio-males-3"]=((df_merge["Males_3rdL"])/df_merge["TOT_M"]).astype(float)
df_merge["ratio-males-2"]=((df_merge["Males_2ndL"])/df_merge["TOT_M"]).astype(float)
df_merge["ratio-males-1"]=((df_merge["Males_1stL"])/df_merge["TOT_M"]).astype(float)

#cal ratios of one lang, two lang and three lang speakers of females
df_merge["ratio-females-3"]=((df_merge["Females_3rdL"])/df_merge["TOT_F"]).astype(float)
df_merge["ratio-females-2"]=((df_merge["Females_2ndL"])/df_merge["TOT_F"]).astype(float)
df_merge["ratio-females-1"]=((df_merge["Females_1stL"])/df_merge["TOT_F"]).astype(float)
df_merge = df_merge[['State Code','Educational_level','ratio-males-3','ratio-females-3','ratio-males-2','ratio-females-2','ratio-males-1','ratio-females-1']]
df_merge
```

```python
#separating the male and female speaker ratios into different dataframes and sorting them by the ratio.
df_3_m=df_merge[['State Code','Educational_level','ratio-males-3']]
df_3_m.sort_values(by=['State Code','ratio-males-3'],ignore_index=True,inplace=True)

df_3_f=df_merge[['State Code','Educational_level','ratio-females-3']]
df_3_f.sort_values(by=['State Code','ratio-females-3'],ignore_index=True,inplace=True)

df_2_m=df_merge[['State Code','Educational_level','ratio-males-2']]
df_2_m.sort_values(by=['State Code','ratio-males-2'],ignore_index=True,inplace=True)

df_2_f=df_merge[['State Code','Educational_level','ratio-females-2']]
df_2_f.sort_values(by=['State Code','ratio-females-2'],ignore_index=True,inplace=True)

df_1_m=df_merge[['State Code','Educational_level','ratio-males-1']]
df_1_m.sort_values(by=['State Code','ratio-males-1'],ignore_index=True,inplace=True)

df_1_f=df_merge[['State Code','Educational_level','ratio-females-1']]
df_1_f.sort_values(by=['State Code','ratio-females-1'],ignore_index=True,inplace=True)   
```

```python
#choosing the highest ratio and corresponding educational level(literacy group)

df_3=df_3_m[5: :6]
df_3["literacy-group-females"]=df_3_f[5: :6]["Educational_level"]
df_3["ratio-females-3"] = df_3_f[5: :6]["ratio-females-3"]
df_3.rename(columns={'State Code':'state/ut','Educational_level':'literacy-group-males'},inplace=True)

df_2=df_2_m[5: :6]
df_2["literacy-group-females"]=df_2_f[5: :6]["Educational_level"]
df_2["ratio-females-2"] = df_2_f[5: :6]["ratio-females-2"]
df_2.rename(columns={'State Code':'state/ut','Educational_level':'literacy-group-males'},inplace=True)

df_1=df_1_m[5: :6]
df_1["literacy-group-females"]=df_1_f[5: :6]["Educational_level"]
df_1["ratio-females-1"] = df_1_f[5: :6]["ratio-females-1"]
df_1.rename(columns={'State Code':'state/ut','Educational_level':'literacy-group-males'},inplace=True)
```

```python
#renaming the columns in the dataframes
df_1.columns=["state/ut","literacy-group-males","ratio-males","literacy-group-females","ratio-females"]
df_2.columns=["state/ut","literacy-group-males","ratio-males","literacy-group-females","ratio-females"]
df_3.columns=["state/ut","literacy-group-males","ratio-males","literacy-group-females","ratio-females"]
```

```python
#writing the output to csv files
outname_1 = 'literacy-gender-c.csv'
outname_2 = 'literacy-gender-b.csv'
outname_3 = 'literacy-gender-a.csv'

outfiles = './Output_Files'
if not os.path.exists(outfiles):
    os.mkdir(outfiles)
outdir=outfiles+'/Q9'
if not os.path.exists(outdir):
    os.mkdir(outdir)
filepath_1 = os.path.join(outdir, outname_1)
filepath_2 = os.path.join(outdir, outname_2)
filepath_3 = os.path.join(outdir, outname_3)

df_1.to_csv(filepath_1,index=False)
df_2.to_csv(filepath_2,index=False)
df_3.to_csv(filepath_3,index=False)
```
