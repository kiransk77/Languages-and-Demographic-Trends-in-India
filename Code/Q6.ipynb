```python
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
#select list of columns required in c8
c8="DDW-0000C-08.xlsx"
cols = pd.read_excel(c8, skiprows=1,nrows=1).columns
cols_d1=['State', 'Area Name', 'Total/', 'Age-group']
cols_d2=cols[6: :3]
cols_c8 = cols_d1+cols_d2.tolist()
```

```python
#load c8
df_c8=pd.read_excel(c8,usecols=cols_c8,skiprows=1,index_col=False)
```

```python
#rename the columns
df_c8.rename(columns={'State':'State Code','Area Name':'Name','Educational level':'Literate without education level',
                     'Unnamed: 18':'Below Primary','Unnamed: 21':'Primary','Unnamed: 24':'Middle',
                      'Unnamed: 27':'Matric/Secondary',
                      'Unnamed: 30':'Higher secondary/Intermediate/Pre-University/Senior secondary',
                     'Unnamed: 33':'Non-technical diploma or certificate not equal to degree',
                    'Unnamed: 36':'Technical diploma or certificate not equal to degree',
                    'Unnamed: 39':'Graduate and above'},inplace=True)

df_c8.drop(['    Total','Literate','Unnamed: 42', 'Unnamed: 45'],axis=1,inplace=True)
df_c8
```

```python
#preprocessing c8 data
df_c8=df_c8[(df_c8['Total/']=='Total')&(df_c8['Age-group']=='All ages')]
df_c8.drop(['Total/','Age-group'],axis=1,inplace=True)
df_c8
```

```python
#adjusting the columns as per those in c19
df_c8['Literate but below primary']=df_c8['Literate without education level']+df_c8['Below Primary']
df_c8['Matric/Secondary but below graduate']= df_c8['Matric/Secondary']+df_c8['Higher secondary/Intermediate/Pre-University/Senior secondary']+df_c8['Non-technical diploma or certificate not equal to degree']+df_c8['Technical diploma or certificate not equal to degree']
df_c8.rename(columns={'Primary':'Primary but below middle','Middle':'Middle but below matric/secondary'},inplace=True)
```

```python
df_c8['State Code']=df_c8['State Code'].astype(int)
df_c8=df_c8[['State Code','Name','Illiterate','Literate but below primary','Primary but below middle','Middle but below matric/secondary','Matric/Secondary but below graduate','Graduate and above']]
```

```python
df_c8 = pd.melt(df_c8,id_vars=['State Code','Name'],
        var_name='Educational_level', value_name='TOT_P')
df_c8.sort_values(by="State Code",inplace=True)
df_c8
```

```python
#load c19
c19="DDW-C19-0000.xlsx"
cols=['State','Area Name','Total/','Educational level','Number speaking third language']

#preprocessing
df_c19=pd.read_excel(c19,usecols=cols,skiprows=[0,2,3,4,5])
df_c19.rename(columns={"Educational level":"Educational_level"},inplace=True)
df_c19
```

```python
#selecting required literacy groups 
exl=['Total','Literate']
df_c19=df_c19[(df_c19['Total/']=='Total')&(~df_c19.Educational_level.isin(exl))]
df_c19.drop(['Total/'],axis=1,inplace=True)
df_c19['State']=df_c19['State'].astype(int)
df_c19
```

```python
#merge c8 and c19
df_merge=pd.merge(df_c8,df_c19,how='inner',left_on=['State Code','Educational_level'],right_on = ['State','Educational_level'])
df_merge.drop(['Name','State'],axis=1,inplace=True)

#percentage of 3 lang speaking 
df_merge["percentage"]=((df_merge["Number speaking third language"] *100)/df_merge["TOT_P"]).astype(float)
df_merge.sort_values(by=["State Code","percentage"],inplace=True)
df_merge
```

```python
#taking highest percentage and corresponding literacy-group in all states
df_final=df_merge[5: :6]
df_final.rename(columns={'State Code':'state/ut','Educational_level':'literacy-group'},inplace=True)
df_final
```

```python
df_final=df_final[['state/ut','literacy-group','percentage']]
```

```python
outname = 'literacy-india.csv'
outfiles = './Output_Files'
if not os.path.exists(outfiles):
    os.mkdir(outfiles)
outdir=outfiles+'/Q6'
if not os.path.exists(outdir):
    os.mkdir(outdir)
filepath = os.path.join(outdir, outname)    

df_final.to_csv(filepath,index=False)
```
