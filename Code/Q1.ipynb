# Q1.Find the percentage of population of India who speaks (a) only one language, (b) exactly two languages, and (c) three languages or more.

```python
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
#Census data loading
census_2011 ="Census2011data.xlsx"
df_census=pd.read_excel(census_2011,usecols= ["State","Level","Name","TRU","TOT_P"])
```

```python
#Census data preprocessing

df_census=df_census[df_census["TRU"]=="Total"]

df_census =df_census[df_census["Level"] != "DISTRICT"]

df_census["State"] = df_census["State"].astype(int)

df_census.rename(columns={'State':'State Code','Name':'State Name'},inplace=True)

df_census = df_census[["State Code","State Name","TOT_P"]]

df_census['State Name']=df_census['State Name'].str.upper()

df_census.reset_index(drop=True,inplace=True)
```

```python
df_census
```

```python
#loading c18
c18="DDW-C18-0000.xlsx"
df_c18=pd.read_excel(c18,skiprows=[0,4,5])
df_c18=df_c18[["State","Area Name","Total/","Age-group","Number speaking second language","Number speaking third language"]]

#c18 preprocessing
df_c18=df_c18[df_c18["Age-group"]=="Total"]
df_c18=df_c18[df_c18["Total/"]=="Total"]

df_c18.rename(columns = {'Area Name':'State Name','State':'State Code'},inplace=True)
df_c18["State Code"] = df_c18["State Code"].astype(int)

df_c18.drop(["Total/","Age-group"],axis=1,inplace=True)
df_c18.reset_index(drop=True,inplace=True)
```

```python
df_c18
```

```python
#merge df_census and df_c18 

df_merge = pd.merge(df_census,df_c18,how='inner',left_on='State Code',right_on = 'State Code')
df_merge
```

```python
#calculate only one language, exactly two languages and Three languages or more speakers

df_merge["Only one language"] = df_merge["TOT_P"]-df_merge["Number speaking second language"]
df_merge["exactly two languages"] = df_merge["Number speaking second language"] - df_merge["Number speaking third language"]
df_merge["Three languages or more"] = df_merge["Number speaking third language"]
df_merge
```

```python
#calculate the percentages for the above 

df_merge["percent-one"] = ((df_merge["Only one language"] *100)/df_merge["TOT_P"]).astype(float).round(2).astype(str)+"%"
df_merge["percent-two"] = ((df_merge["exactly two languages"] *100)/df_merge["TOT_P"]).astype(float).round(2).astype(str)+"%"
df_merge["percent-three"] = ((df_merge["Three languages or more"] *100)/df_merge["TOT_P"]).astype(float).round(2).astype(str)+"%"
df_merge
```

```python
#select required attributes 

df_final=df_merge[["State Code","percent-one","percent-two","percent-three"]]
```

```python
df_final
```

```python
#write to output csv file 

outname = 'percent-india.csv'
outfiles = './Output_Files'
if not os.path.exists(outfiles):
    os.mkdir(outfiles)
outdir=outfiles+'/Q1'
if not os.path.exists(outdir):
    os.mkdir(outdir)
filepath = os.path.join(outdir, outname)    

df_final.to_csv(filepath,index=False)
```
